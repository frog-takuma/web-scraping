### 1-詳細ページへのＵＲＬリスト収集

```python
driver.get("https://doctorsfile.jp/search/pv1543/page/1")
list_links = []\

while True:
    
    add_links = [i.find_elements_by_class_name('result__name')[0].get_attribute('href') for i in driver.find_elements_by_class_name('result')]
    list_links.extend(add_links)

    try:
        bottan = driver.find_elements_by_class_name('pagination__next')[0] # 
        bottan.click() #”次へ”クリック

    except:
        break
```

### 2-詳細ページへに移動

```python
URL_page = []
name = []
hospital_department =[]
tel = []
address = []
point = []


for i,d in enumerate(list_links):
    driver.get(d)
    print("\r"+"読み込み中-{}/{}ページ".format(i,len(list_links)),end="")

    makeUrl = "https://msearch.gsi.go.jp/address-search/AddressSearch?q="
    s_quote = urllib.parse.quote(driver.find_elements_by_class_name('table-clinic__data')[-1].text.split(" ")[1])
    response = requests.get(makeUrl + s_quote)
    #print(response.json()[0]["geometry"]["coordinates"])
       

    URL_page.append(d)
    name.append(driver.find_elements_by_class_name('table-clinic__data')[1].text)
    hospital_department.append(driver.find_elements_by_class_name('table-clinic__data')[2].text)
    tel.append(driver.find_elements_by_class_name('table-clinic__data')[-2].text)
    address.append(driver.find_elements_by_class_name('table-clinic__data')[-1].text)
    point.append(response.json()[0]["geometry"]["coordinates"])

    
    """
    for i in driver.find_elements_by_class_name('table-clinic__data'):
        print("#########################################")
        print(i.text)
        
    """
```

### 3-CSVに保存

```python
data = {"URL":URL_page,
        "hospital_name":name,
        "hospital_departoment":hospital_department,
        "tel":tel,
        "address":address,
        "address_point":point}

df = pd.DataFrame(data)
df.to_csv('save_csv_data\\hospital_data.csv', index=False, encoding="shift-jis")
    
```
