```python
#参照 https://ai-inter1.com/python-selenium/\

import os
from selenium import webdriver
from time import sleep

#google chromeを開く
driver = webdriver.Chrome("C:\\Users\\kashi\\OneDrive\\ドキュメント\\A2\\python-proguram\\web-scra\\chromedrive\\chromedriver.exe")
```

```python
#URL下記のアドレスを入力して開く
driver.get('https://www.google.co.jp')

#上記で検索したHTMLファイル内の”ｑ”を指定する
search_bar = driver.find_element_by_name("q")

#上記の”ｑ”は検索欄→”python”と入力
search_bar.send_keys("python")

#検索を実行
search_bar.submit()
```

```python
# Xpathとは　https://www.octoparse.jp/blog/xpath-introduction/#div2

for elem_h3 in driver.find_elements_by_xpath('//a/h3'): # // → 以下略称 
    elem_a = elem_h3.find_element_by_xpath('..') #h3の親タグ ".." → 親タグ(直前のタグ)
    print(elem_h3.text)
    print(elem_a.get_attribute('href')) # https://www.seleniumqref.com/api/python/element_infoget/Python_get_attribute.html
   
```
